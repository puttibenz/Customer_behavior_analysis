```python
import pandas as pd

file_path = '/content/drive/My Drive/Colab Notebooks/customer_shopping_behavior.csv'
df = pd.read_csv(file_path)

display(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Review Rating')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
#create a column age _group
labels = ['Young Adult','Adult','Middle-Age','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df[['age','age_group']].head()
```

```python
#create a column purchase_frequency_days
mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(mapping)
df[['frequency_of_purchases','purchase_frequency_days']].head()
```

```python
df[['discount_applied','promo_code_used']]
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used',axis=1)
```
